```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
#DataFrame
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
# Display some rows of the DataFrame
print(df.head())
```

```python
# Box Plot
plt.figure(figsize=(12, 8))
df.boxplot(column=iris.feature_names, by='species')
plt.suptitle('')
plt.title('Box Plot of Iris Features by Species')
plt.show()
```

```python
# Bar Plot
plt.figure(figsize=(12, 8))
df.groupby('species').mean().plot(kind='bar')
plt.title('Bar Plot of Average Iris Features by Species')
plt.ylabel('Mean Feature Value')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(12, 8))
for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(subset[iris.feature_names[0]], subset[iris.feature_names[1]], label=species)

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Scatter Plot of Iris Dataset')
plt.legend()
plt.show()
```
